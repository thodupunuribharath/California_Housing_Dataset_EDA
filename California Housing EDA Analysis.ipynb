## California Housing Data set: 

### This dataset having the overall data of California district and also the summary of various statistics from 1900 census collection.

### Data Understanding: It has the overall 20,640 records with 10 columns as features.

### --> Data basically describes about the california state which is ultimately located in USA.

### --> Each feature in the data set can be explained as below.

1. longitude:A measure of how far west a house is; a more negative value is farther west  
(Numerical :- Contineous)

2. latitude:A measure of how far north a house is; a higher value is farther north
(Numerical :- Contineous)

3. housingMedianAge:Median age of a house within a block; a lower number is a newer building
(Numerical :- Contineous)

4. totalRooms:Total number of rooms within a block
(Numerical :- Discrete)

5. totalBedrooms:Total number of bedrooms within a block
(Numerical :- Discrete)

6. population:Total number of people residing within a block
(Numerical :- Discrete)

7. households:Total number of households, a group of people residing within a home unit, for a block
(Numerical :- Discrete)

8. medianIncome:Median income for households within a block of houses (measured in tens of thousands of US Dollars)
(Numerical :- Continuous)

9. medianHouseValue:Median house value for households within a block (measured in US Dollars)
(Numerical :- Discrete)

Outcome: This is basically used to predict the california house prices how it vary usuallay based on the future by several factors acting on it.

### 10. Please create a new column named total_bedroom_size. If size 10 or less. It should be quoted as small. If the size is 11 or more but 1000 less, it should be medium, otherwise, it should be large.

In this we have required majorly the three conditions as below.
1. <= 10 are small
2. <= 1000 are Medium
3. >1000 are Large

for the specific coloumn called "Total Bed room size"


```python
conditions=[
    
    (housing["total_bedrooms"]<=10),       #condition for small
    
    (housing["total_bedrooms"]<=1000),    #condition for medium
    
    (housing["total_bedrooms"]>1000)      #condition for large
]
values=["Small","Medium","Large"]    #assigning condition

housing["total_bedroom_size"]= np.select(conditions,values)  #selection of values

housing  #priting the new coloumn
```

A New coloumn has been created for  the Total Bed Room size
